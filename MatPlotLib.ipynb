## Ploting between two varible x and y !! 

```python
x = [2,3,5,6,7]
y = [4,5,7,6,8]
plt.plot(x,y)
```

```python
# 1st method 
fig = plt.figure() #create a figure
ax = fig.add_subplot() #add some axes
plt.show


```

```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3ld method 
fig, ax = plt.subplots()
ax.plot(x,y);
```

# MatplotLib Workflow

```python
# 0. Import statement
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x = [1,3,4,3,5]
y = [4,5,6,7,5]

#2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # (width,hight)

#3. Plot data
ax.plot(x,y)

#4. customise plot
ax.set(title="Simple plot",
       xlabel="X-axis",
       ylabel="Y-axis"
      )

#5. save and show( save and show whole figure)
fig.savefig("images/sample-plot.png")
```
