# Importing MatplotLib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.plot()
```

## Ploting between two varible x and y !! 

```python
x = [2,3,5,6,7]
y = [4,5,7,6,8]
plt.plot(x,y)
```

```python
# 1st method 
fig = plt.figure() #create a figure
ax = fig.add_subplot() #add some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3ld method 
fig, ax = plt.subplots()
ax.plot(x,y);
```
